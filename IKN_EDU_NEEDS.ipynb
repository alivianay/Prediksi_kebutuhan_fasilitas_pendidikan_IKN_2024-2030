```python
import yfinance as yf

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from statsmodels.tsa.stattools import adfuller

from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf

from sklearn.metrics import mean_squared_error

import warnings
warnings.filterwarnings("ignore")
```

# Import Data


```python
df = yf.download("MSFT", start="2024-01-01", end="2030-01-01")
print(df.tail())
```

```python
df["years"] = df.index.strftime("%V")
df = df.groupby(['years']).mean()
df.head()
```

# Time Series Plot

```python
series = df["Adj Close"]
```

```python
plt.plot(df.index, series)
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Stationarity

ACF and PACF assume stationarity of the underlying time series.
Stationarity can be checked by performing an Augmented Dickey-Fuller (ADF) test:

p-value > 0.05: Fail to reject the null hypothesis (H0), the data has a unit root and is non-stationary.

p-value <= 0.05: Reject the null hypothesis (H0), the data does not have a unit root and is stationary.

```python
adfuller(series)[0:2]
```

# Differencing

diff(1) = y(t) - y(t-1)

```python
def difference(dataset, interval=1):
 diff = list()
 for i in range(interval, len(dataset)):
  value = dataset[i] - dataset[i - interval]
  diff.append(value)
 return(diff)
```

```python
diff = series.diff()
diff.head()
```

```python
diff = diff.dropna()
```

```python
plt.plot(df.index[1:], diff)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
adfuller(diff)[0:2]
```

# ACF and PACF

```python
f, ax = plt.subplots(nrows=2, ncols=1, figsize=(6,6))
plot_acf(diff, lags=min(20, len(diff)//2), ax=ax[0])
plot_pacf(diff, lags=min(20, len(diff)//2), ax=ax[1], method='ols')

plt.tight_layout()
plt.show()
```

# ARIMA model

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(series, order=(1,1,0)).fit()
print(model.summary())
```

```python
print(model.forecast())
```

```python
# line plot of residuals
residuals = pd.DataFrame(model.resid)
residuals.plot()
plt.show()
# density plot of residuals
residuals.plot(kind='kde')
plt.show()
# summary stats of residuals
print(residuals.describe())
```

```python
df_fore = yf.download("MSFT", start="2024-01-01", end="2030-01-01")
print(df_fore.tail())
df_fore["years"] = df_fore.index.strftime("%V")
df_fore = df_fore.groupby(['years']).mean()
df_fore.tail()
```

```python
fore = df_fore['Adj Close']
```

```python
plt.plot(df_fore.index, fore)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
series_f = series
predictions = list()
# walk-forward validation
for t in range(len(fore)):
  model = ARIMA(series_f, order=(1,1,1)).fit()
  output = model.forecast()
  predictions.append(output)
  obs = fore[t]
  series_f = series_f.append(pd.Series([obs], index=[df_fore.index[t]]))
  print('predicted=%f, expected=%f' % (output, obs))
# evaluate forecasts
rmse = np.sqrt(mean_squared_error(fore, predictions))
print('Test RMSE: %.3f' % rmse)
# plot forecasts against actual outcomes
plt.plot(df_fore.index, fore)
plt.plot(df_fore.index, predictions, color='red')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# NN for Time Series

Time Series Plot

```python
series = df["Adj Close"]
```

```python
plt.plot(df.index, series)
plt.xticks(rotation=90, ha='right')
plt.show()
```

Train Val

```python
series.plot()
```

```python
split_date = 7 #batasnya data ke berapa
df1 = series
train = df1.iloc[:split_date]
test = df1.iloc[split_date:]
plt.figure()
plt.xlim(0,51)
plt.plot(train)
plt.plot(test)
plt.legend(['train', 'test'])
print(len(train), len(test))
```

Lag for Input Variable

```python
lag1 = series.shift(1)
```

```python
lag1
```

```python
lag2 = lag1.shift(1)
lag2
```

```python
nn_data = pd.DataFrame({'series':series, 'lag1':lag1, 'lag2':lag2})
nn_data
```

```python
nn_data = nn_data.dropna()
```

```python
nn_data
```

```python
split_date = 7
train = nn_data.iloc[:split_date]
test = nn_data.iloc[split_date:]

from scipy.ndimage import shift
X_train = np.asarray(train[['lag1','lag2']])
y_train = np.asarray(train['series'])
X_train.shape, y_train.shape
```

 Modeling

```python
from keras.models import Sequential
from keras.layers import Dense

nn_model = Sequential()
nn_model.add(Dense(12, input_dim=2, activation='relu'))
nn_model.add(Dense(1))
nn_model.compile(loss='mean_squared_error', optimizer='adam')
history = nn_model.fit(X_train, y_train, epochs=100, batch_size=1, verbose=1, shuffle=False)
```

Testing

```python
from sklearn.metrics import r2_score
```

```python
X_test = np.asarray(test[['lag1','lag2']])
y_test = np.asarray(test['series'])

y_pred_test_nn = nn_model.predict(X_test)
y_train_pred_nn = nn_model.predict(X_train)
print("The R2 score on the Train set is:\t{:0.3f}".format(r2_score(y_train, y_train_pred_nn)))
print("The R2 score on the Test set is:\t{:0.3f}".format(r2_score(y_test, y_pred_test_nn)))
```

```python
plt.plot(test.index, y_test)
plt.plot(test.index, y_pred_test_nn, color='red')
plt.xticks(rotation=45, ha='right')
plt.show()
```
